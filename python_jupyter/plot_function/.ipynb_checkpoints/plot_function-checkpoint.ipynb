This is a simple test of the Jupyter notebook environment. I will be plotting the function $f(x)=x^2$.

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def fun(x):
    """ Define the function """
    return x**2
```

Endpoints and resolution:

```python
start = -2.0
stop = 2.0
res = 1000
```

The $x$ and $y$ values:

```python
x = np.linspace(start, stop, res)
y = fun(x)
```

Plot the results:

```python
plt.rc('text', usetex=True)
plt.plot(x,y)
plt.legend([r"$f(x) = x^2$"])
plt.title("Plot Title")
plt.savefig("plot.png")
plt.show()
```
